```python
###Creating two new columns for average minutes per game and goals per game 
df['AvgMinutesPerGame'] = (df['Min']/df['MP']).astype(int)
df['AvgGoalsPerGame']=(df['Gls']/df['MP']).astype(float)
df.head()
```

```python
#Total Goals, Total non penalty goals, total penalty goals and total non pk gls 
Total_Goals = df['Gls'].sum()
print(f'Total Goals in UCL 2021-2022 is: {Total_Goals}')
Total_Non_Pen_Goals = df['G-PK'].sum()
print(f'Total Non Penalty Goals in UCL 2021-2022 is: {Total_Non_Pen_Goals}')
Total_PKs = df['PKatt'].sum()
print(f'Total Penalty Kicks Attempted in UCL 2021-2022 is: {Total_PKs}')
Total_PK_Gls = df['PK'].sum()
print(f'Total Penalty Kicks Actually Scored in UCL 2021-2022 is: {Total_PK_Gls}')
```

```python
#Visualization of said data
#Breaking down now to percentage of goals as opposed to pk
plt.figure(figsize=(14, 8))
plt_non_scored_PK = df['PKatt'].sum() - Total_PK_Gls 
data = [plt_non_scored_PK, Total_PK_Gls]
labels = ['Penalties Missed','Penalties Scored']
color = sns.color_palette("Paired")
plt.pie(data, labels = labels, colors = color, autopct=lambda p:f'{p:.2f}%, {p*sum(data)/100 :.0f} Goals')
plt.show()
```

```python
#Viz 
# percentage of pks attempted and actually socred
plt.figure(figsize=(14, 8))
plt_non_PK_scored = df['Gls'].sum() - Total_PK_Gls 
data = [plt_non_PK_scored, Total_PK_Gls]
labels = ['Non-PK Goals Scored','Penalties Scored']
color = sns.color_palette("Set2")
plt.pie(data, labels = labels, colors = color, autopct=lambda p:f'{p:.2f}%, {p*sum(data)/100 :.0f} Goals')
plt.show()
```

```python
#Distributing players based on age 
sns.displot(df, x="Age", color = "teal")
```

```python
sns.boxplot(df['Age'])
```

```python
#More data expl. on age 
under_20 = df[df['Age'] <= 20]
aged20_25 = df[(df['Age'] > 20) & (df['Age'] <=25)]
aged25_30 = df[(df['Age'] > 25) & (df['Age'] <=30)]
over30 = df[df['Age'] >30]
```

```python
#Now to visualize the age distribution 
ages = np.array([under_20['Player'].count(), aged20_25['Player'].count(), aged25_30['Player'].count(), over30['Player'].count()])
mylabels = ['20 and Younger','Between 21 and 25', 'Between 26 and 30','Over 30']
plt.figure(figsize = (20,8))
plt.title('Age Distribution in UCL 2021-2022', fontsize = 18)
plt.pie(ages, labels = mylabels, autopct=lambda p:f'{p:.2f}%, {p*sum(ages)/100 :.0f} Players')

```

```python
#Squads with most under 20 player 
plt.figure(figsize = (16,8))
players_under_20 = df[df['Age'] < 20]
players_under_20['Squad'].value_counts().plot(kind = 'bar', title = 'Clubs with most U-20 Players in UCL 2021-2022' ,
    color = sns.color_palette('magma'))
```

```python
#Now Moving to Assists/Goals ---- Total Assists Per Club 
Assists_per_club = pd.DataFrame(df.groupby("Squad", as_index = False)['Ast'].sum())
sns.set_theme(style= "whitegrid", color_codes= True)
amp = sns.barplot(x='Squad', y ='Ast', data = Assists_per_club.sort_values(by = 'Ast'), palette='crest')
amp.set_xlabel('Squad', fontsize = 20)
amp.set_ylabel('Assists', fontsize = 20)
plt.rcParams['figure.figsize'] = (60,40)
plt.title("Assists per Club for UCL 2021-2022", fontsize = 40)
```

```python
#Top 15 Players with Most Assists 
top_15 = df[['Player', 'Squad', 'Ast', 'MP']].nlargest(n=15, columns = 'Ast')
top_15
```

```python
#Total Goals Per Club 
goals_per_club = pd.DataFrame(df.groupby("Squad", as_index = False)['Gls'].sum())
sns.set_theme(style= "whitegrid", color_codes= True)
amp = sns.barplot(x='Squad', y ='Gls', data = goals_per_club.sort_values(by = 'Gls'), palette='viridis')
amp.set_xlabel('Squad', fontsize = 20)
amp.set_ylabel('Goals', fontsize = 20)
plt.rcParams['figure.figsize'] = (60,40)
plt.title("Goals per Club for UCL 2021-2022", fontsize = 40)
```

```python
#Top 15 Players with Most Goals
top_15 = df[['Player', 'Squad', 'Gls', 'MP']].nlargest(n=15, columns = 'Gls')
top_15
```

```python
#Trying to figure out how many of the goals actually came with a provided assists 
plt.figure(figsize =(20,8))
assists = df['Ast'].sum()
data = [Total_Goals- assists, assists]
labels = ['Goals w/ Asissts','Goals w/out Asissts']
color = sns.color_palette('Paired')
plt.pie(data, labels = labels, colors = color, autopct=lambda p:f'{p:.2f}%' )
plt.show()
```
